```python
# Import pustaka
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Menetapkan seed supaya data rawak adalah konsisten (pilihan)
np.random.seed(42)

# Menjana matriks data rawak bersaiz 10 x 3
data = np.random.randint(1, 100, size=(10, 3))

# Menukar matriks kepada DataFrame
df = pd.DataFrame(data, columns=['A', 'B', 'C'])

# Memaparkan DataFrame
print("DataFrame:")
print(df)

# Melukis graf garisan
plt.figure(figsize=(8, 5))
plt.plot(df.index, df['A'], marker='o', label='A')
plt.plot(df.index, df['B'], marker='s', label='B')
plt.plot(df.index, df['C'], marker='^', label='C')

# Menambah tajuk dan label
plt.title('Graf Garisan Data Rawak')
plt.xlabel('Indeks Data')
plt.ylabel('Nilai')
plt.grid(True)
plt.legend()

# Memaparkan graf
plt.show()
```
